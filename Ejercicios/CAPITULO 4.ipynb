### Aprendizaje Supervisado

#### Clasificación
El aprendizaje supervisado de clasificación es una técnica en el campo de la inteligencia y el aprendizaje automático. Se utiliza para predecir a que clase o categoria pertenece un nuevo conjunto de datos, basándose en el conocimiento adquirido de datos previamente etiquetados.
En este tipo de aprendizaje, el algoritmo recibe un conjunto de datos de entrenamiento que consiste en entradas (variables independientes) junto con las etiquetas de salida correspondientes (variables dependientes o clases ). El algoritmo aprende a mapear las entradas a las etiquetas de salida observando los datos de entranamiento y ajustando sus parámetros.
Un vez entrenando el modelo, se puede utilizar para predecir la clase de nuevas entradas que no formaban parte del conjunto de entrenamiento. Por ejemplo, supongamos que queremos clasificar correos electronicos como spam, o no spam. Para ello, recopilamos un conjunto de datos de correso electronicos previamente etiquetados como spam o no spam, junto con caracteríticas como la frecuencia de ciertas palabras. Utilizamos este conjunto de datos para entrenar un modelo de aprendizaje supervisado de clasificación. Una vez entrenado, el modelo puede predecir si un nuevo correo electronico es spam o no spam en función de sus características.

Ejemplos en los que se utiliza la clasificación
- Detección de fraudes.
- Detección de Spam.
- Diagnósticos.
- Clasificación de imágenes.
- Retención de clientes.

#### Regresión
La regresión es un método estadístico utilizado para modelar y analizar la relación entre una variable dependiente (también conocida como de respuesta o variable objetivo) y una o más variables independientes (también conocidas como variables predictoras o variables de entrada).
El objetivo principal de la regresión es comprender cómo una o más variables independientes afectan a la variable dependiente y utilizar esta relación para predecir o estimar valores futuros de la variable dependiente.
En la regresión, se busca ajustar una función matemática a los datos observados que minimice la discrepancia entre los valores observados y los valores predichos por el modelo.
Existen varios tipos de regresión, entre los que se incluyen la regresión lineal (donde se asume una relación lineal entre las variables), la regresión logística (utilizada para problemas de clasificación binaria), la regresión polinomial(donde se permite una relación no lineal entre las variables) y muchos más.
Por ejemplos, en el caso de la regresion lineal simple, se busca ajustar una linea recta a los datos para modelar la relación entre una variable independiente y una variable dependiente. Esta línea puede ser utilizada para predecir el valor de la variable dependeinte para nuevos valores de la variables independiente.
En resumen, la regresión es una herramienta estadística utilizada para modelar la relación entrevariables y predecir valores futuros basados en esta relación.

Ejemplos en los que se utiliza la regresión
- Predicciones.
- Pronósticos.
- Optimización de procesos.
- Nuevas perspectivas.

## Tratamiento de datos finales

### Trabajar con nulos

```python
datos_inmuebles.head()
```

```python
datos_inmuebles.address_zone.value_counts()
```

```python
datos_inmuebles["address_zone"] = datos_inmuebles["address_zone"].replace("", np.nan)
```

```python
datos_inmuebles.address_zone.isnull().sum()
```

```python
datos_inmuebles.head()
```

```python
# Consultamos la tabla 'datos_inmuebles' para encontrar columnas con valores nulos.
# Seleccionamos las filas donde la columna 'address_zone' no es nula utilizando 'isna()'.
# Utilizamos el operador '~' para obtener todas las filas donde 'address_zone' no es nulo.
# Esto filtra las filas con valores nulos en 'address_zone'.

# Luego de obtener las filas con 'address_zone' no nulo, eliminamos las filas duplicadas 
# basadas en la columna 'address_neighborhood' utilizando 'drop_duplicates'.
# Esto nos da una lista de registros únicos donde cada uno tiene un valor único en 'address_neighborhood'.

# Finalmente, convertimos este conjunto de registros únicos en un diccionario con 'to_dict("records")'.
dici = datos_inmuebles[~datos_inmuebles["address_zone"].isna()].drop_duplicates(subset=[
    "address_neighborhood"]).to_dict("records")
```

```python
print(dici)
```

```python
dic_zonas = {dic["address_neighborhood"]: dic["address_zone"] for dic in dici}
```

```python
print(dic_zonas)
```

```python
for barrio, zona in dic_zonas.items():
    datos_inmuebles.loc[datos_inmuebles["address_neighborhood"] == barrio, "address_zone"] = zona
```
