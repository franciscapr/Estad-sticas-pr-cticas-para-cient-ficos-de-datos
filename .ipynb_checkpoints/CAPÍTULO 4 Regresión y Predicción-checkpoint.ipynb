# **Regresión y Predicción**

Quizás el objetivo más común en estadística sea responder a la pregunta "¿Está la variable X (o más probablemente, $X1, ..., Xp$) asociada con una variable $Y$, y si es así, cuál es la relación y podemos usarla para predecir $Y$?"
Ningún lugar muestra una conexión más fuerte entre estadística y ciencia de datos que en el ámbito de la predicción, específicamente, la predicción de una variable de resultado (objetivo) basada en los valores de otras variables "predictoras". Este proceso de entrenar un modelo con datos donde se conoce el resultado, para su posterior aplicación a datos donde el resultado no se conoce, se denomina aprendizaje supervisado. Otra conexión importante entre la ciencia de datos y la estadística se encuentra en el área de detección de anomalías, donde los diagnósticos de regresión originalmente destinados al análisis de datos y la mejora del modelo de regresión pueden utilizarse para detectar registros inusuales.

La regresión y la predicción son herramientas estadísticas que nos ayudan a entender y predecir relaciones entre variables.
- `Regresión`: La regresión es como tratar de dibujar una línea o una curva que mejor se ajuste a nuestros datos. Imagina que tienes puntos dispersos en un gráfico. La regresión intenta encontrar la mejor forma de trazar una línea o curva a través de esos puntos para describir la relación entre dos variables. Por ejemplo, podrías querer saber cómo el precio de una casa (variable dependiente) está relacionado con su tamaño (variable independiente). La regresión te ayudaría a encontrar cómo estos dos factores están relacionados.
- `Predicción`: Una vez que hemos encontrado esa línea o curva que mejor se ajusta a nuestros datos, podemos usarla para hacer predicciones sobre lo que podría suceder en el futuro. Por ejemplo, si hemos estudiado la relación entre el tamaño de una casa y su precio, podríamos predecri el precio de una casa nueva basándonos en su tamaño. Esto es útil en muchos campos, como en la economía para prever el crecimiento del PIB o en medicina para predecir la probabilidad de que una enfermedad progrese.
- La regresión nos ayuda a entender cómo se relacionan dos variables, miesntras que la predicción nos permite utilizar esa relación para hacer estimaciones sobre eventos futuros.

# **Regresión lineal simple**

La `regresión lineal simple` ofrece un modelo de la relación entre la magnitud de una variable y la de una segunda variable, por ejemplo, a medida que $X$ aumenta, $Y$ también aumenta. O a medida que $X$ aumenta, $Y$ disminuye. La `correlación` es otra forma de medir cómo dos variables están relacionadas; consulta la sección "Correlación" en la página 30. La diferencia radica en que, mientras que la correlación mide la fuerza de la asociación entre dos variables, la regresión cuantifica la naturaleza de la relación.

### Ejemplo

- Supongamos que queremos entender cómo el tiempo de estudio afecta las calificaciones de una estudiante en un examen.
- Datos: Tenemos datos de 10 estudiante que muestran cuántas horas estudiaron $X$ y sus calificaciones en el examen $Y$.

        Horas de estudio (X)	Calificación en el examen (Y)
                 2	                          60
                 3	                          65
                 4	                          70
                 5	                          75
                 6	                          80
                 7	                          85
                 8	                          90
                 9	                          95
                 10	                          100
                 11	                          105

- Gráfico: Podemos graficar estos datos en un gráfico de dispersión, donde el eje $X$ representa la horas y el eje $Y$ representa las calificaciones en el examen. Varemos que a medida que aumenta el tiempo de estudio, las calificaciones también tienden a aumentar.
- Regresión lineal: Aplicamos la regresión lineal simple para encontrar la línea que mejor se ajuste a estos datos. Esta línea nos mostrará cómo las calificaciones cambian en promedio con cada hora adicional de estudio, imagina una línea recta que atraviesa todos los puntso en el gráfico, tratando de estar lo más cerca posible de todos ellos.
- Interpretación de la línea e egresión: Si la línea de regresión tiene una pendiente positiva, significa que a medida que aumenta el tiempo de estudio $X$, las calificaciones $Y$ también aumenta. Si la pendiente es negativa, entoncs a medida que aumenta el tiempo de estudio, las calificaciones diminuyen.
- Predicción: Usando esta línea de regresión podemos hacer predicciones sobre las calificaciones de un estudiante basadas en cuánto tiempo estudia. Por ejemplo, si un estudiante estudia durante 7 hotas, podemos predecir que obtendrá una calificacion alrededor de 85.

- La regresión lineal simple es una método estadistico utilizado para modelar la relación entre una variable independiente ($X$) y una variable dependiente ($Y$). Se asume que esta relación puede ser aproximada por una linea recta en un plano cartesiano.
- Cuando se dice que la regresión lineal simple ofrece un modelo de la relación entre dos variables, significa que trata de explicar cómo cambia una variable ($Y$) en función de los cambios en otra variable ($X$).
- Por ejemplo, si estamos tratando de predecir la temperatura ($Y$) en función de la cantidad de horas de sol ($X$), podríamos usar la regresión lineal simple para modelar cómo la temperatura cambia cuando la cantidad de horas de sol varia.
- La relación entre $X$ e $Y$ puede ser de diferentes tipos: positiva, negativa, nula. En una relación positiva, a medida que $X$ aumenta, $Y$ también aumenta. En una relación negativa, a medida que $X$ aumenta, $Y$ disminuye. $Y$ en una relación nula, no hay una relación evidente entre $X$ e $Y$.
- Es importante destacar que la regresión lineal simple se basa en la suposición de que la relación entre las variables puede ser aproximada por una linea recta. Esta línea recta se ajusta a los daros de la mejor manera posible para minimizar la diferencia entre los valores observados de $Y$ y los valores predicho por el modelo.
- En resumen, la regresión lineal simple es una herramienta estadíctica que nos permite entender y modelar la relación entre dos variables, cuantificando la naturaleza de esa relación y facilitando la predicción de los valores de una variable en función de los valores de la otra variable.

### Términos clave para la regresión lineal simple

- **Respuesta** `Response`
  - La variable que intentamos predecir.
  - Sinónimos: variable dependiente, variable Y, objetivo (`target`), resultado.

- **Variable independiente**
  - La variable utilizada para predecir la respuesta.
  - Sinónimos: variable X, característica (`feature`), atributo, predictor.

- **Registro**
  - El vector de valores predictores y resultados para un individuo o caso específico.
  - Sinónimos: fila, caso, instancia, ejemplo.

- **Intercepto**
  - El intercepto de la línea de regresión, es decir, el valor predicho cuando $X=0$.
  - Sinónimos: $b0$, $β0$.

- **Coeficiente de regresión**
  - La pendiente de la línea de regresión.
  - Sinónimos: pendiente, $b1$, $β1$, estimaciones de parámetros, pesos.

- **Valores ajustados**
  - Las estimaciones $\bar{Yi}$ obtenidas de la línea de regresión.
  - Sinónimo: valores predichos.

- **Residuos**
  - La diferencia entre los valores observados y los valores ajustados.
  - Sinónimo: errores.

- **Mínimos cuadrados**
  - El método de ajustar una regresión minimizando la suma de los residuos al cuadrado.
  - Sinónimos: mínimos cuadrados ordinarios, OLS (por sus siglas en inglés).

# **La Ecuación de Regresión**

La `regresión lineal simple` estima cuánto cambiará $Y$ cuando $X$ cambie por una cantidad determinada. Con el `coeficiente de correlación`, las variables $X$ e $Y$ son intercambiables. Con la regresión, intentamos predecir la variable $Y$ a partir de $X$ utilizando una relación lineal (es decir, una línea):

$$Y = b_0 + b_1 X$$ 

Leemos esto como "$Y$ es igual a $b_1$ veces $X$, más una constante $b_0$". El símbolo $b_0$ se conoce como la `intersección` (o constante), y el símbolo $b_1$ como la pendiente para $X$. Ambos aparecen en la salida de R como coeficientes, aunque en el uso general, el término coeficiente a menudo se reserva para $b_1$. La variable $Y$ se conoce como la respuesta o variable dependiente, ya que depende de $X$. La variable $X$ se conoce como el predictor o variable independiente. La comunidad de aprendizaje automático tiende a utilizar otros términos, llamando a $Y$ el objetivo y a $X$ un vector de características. A lo largo de este libro, utilizaremos los términos predictor `"predictor"` y característica `"feature"` de manera intercambiable.

Considera el gráfico de dispersión en la Figura 4-1 que muestra el número de años que un trabajador estuvo expuesto al polvo de algodón (Exposición) versus una medida de capacidad pulmonar (PEFR o "tasa de flujo espiratorio máximo"). ¿Cómo está relacionado PEFR con la Exposición? Es difícil decirlo solo con la imagen.

1. La ecuación de regresión lineal simple: La ecuación fundamental de la regresión lineal simple se presenta como $Y = B_0 + B_1X$. En esta ecuación, $Y$ es la variable dependiente (también llamada respuesta "response"), $X$ es la variable independiente (también llamada "predictor" o "feature"), $b_0$ es la intersección o constante, y $b_1$ es la pendiente para $X$.
2. Interpretación de los coeficientes: $b_0$ representa el valor de $Y$ cuando $X$ es igual cero. $b_1$ indica cuánto cambiaá $Y$ por cada unidad de cambio en $X$. Por lo tanto, la pendiente $b_1$ cuantifica la fuerza y dirección de la relación entre $X$ e $Y$.
3. Nomenclatura alternativa: En la comunidad de parendizaje automático, $Y$ se conoce como el objetivo y $X$ se denomina un vectori de caracteristicas. A lo laergo del libro, los términos "predictor" y "caracteristica" se utilizan de manera intercambiable para referirse a $X$.

La `regresión lineal simple` intenta encontrar la "mejor" línea para predecir la respuesta PEFR $Y$ como función de la variable predictor Exposure $X$:

$$PEFR = b_0 + b_1 \cdot \text{Exposure}$$

La función lm en R se puede utilizar para ajustar una regresión lineal:

    model <- lm(PEFR ~ Exposure, data=lung)

Correcto, en este contexto, "lm" significa "modelo lineal", y el símbolo "~" denota que PEFR es predicho por Exposure. Con esta definición de modelo, la intersección se incluye automáticamente y se ajusta. Si deseas excluir la intersección del modelo, debes escribir la definición del modelo de la siguiente manera:

    PEFR ~ Exposure - 1

Imprimir el objeto del modelo produce la siguiente salida:

    Call:
    lm(formula = PEFR ~ Exposure, data = lung)
    Coefficients:
    (Intercept) Exposure
     424.583 -4.185

La `intersección`, o $b_0$, es 424.583 y se puede interpretar como la PEFR predicha para un trabajador con cero años de exposición. El `coeficiente de regresión`, o $b_1$ , se puede interpretar de la siguiente manera: por cada año adicional que un trabajador esté expuesto al polvo de algodón, la medición de PEFR del trabajador se reduce en -4.185.

En Python, podemos usar `LinearRegression` del paquete `scikit-learn`. (el paquete statsmodels tiene una implementación de regresión lineal que es más similar a R (sm.OLS); la usaremos más adelante en este capítulo):

```python
# Definimos las variables
predictors = ["Exposure"]    # Variable independiente predictora "X"
outcome = "PEFR"    # Variable dependiente "Y"

# Definimos el modelo que utilizaremos en este caso LinearRegression()
model = LinearRegression()    # Creamos un objeto de modelo de regresión lineal y lo asignamos a model
model.fit(lung[predictors], lung[outcome])    # Ajustamos el modelo utilizando los datos de lung[predictors] 
# como variables independientes y lung[outcome] como variable dependiente, pasamos estos argumentos

# Imprimimos los coeficientes del modelo
print(f"Intercept: {model.intercept_:.3f}")    # Imprimimos el término independiente (intercepto) del modelo ajustado
print(f"Coefficiente Exposure: {model.coef_[0]:.3f}")    # Imprimimos el coeficiente asociado a la variable 
#"Exposure" del modelo ajustado
```

La línea de regresión de este modelo se muestra en la Figura 4-2.

```python
fig, ax = plt.subplots(figsize=(4, 4))
ax.set_xlim(0, 23)
ax.set_ylim(295, 450)
ax.set_xlabel("Exposure")
ax.set_ylabel("PEFR")
ax.plot((0, 23), model.predict(pd.DataFrame({"Exposure": [0, 23]})))
ax.text(0.4, model.intercept_, r"$b_0$", size="larger")

x = pd.DataFrame({"Exposure": [7.5,17.5]})
y = model.predict(x)
ax.plot((7.5, 7.5, 17.5), (y[0], y[1], y[1]), "--")
ax.text(5, np.mean(y), r"$\Delta Y$", size="larger")
ax.text(12, y[1] - 10, r"$\Delta X$", size="larger")
ax.text(12, 390, r"$b_1 = \frac{\Delta Y}{\Delta X}$", size="larger")

plt.tight_layout()
plt.show()
```

**Figura 4-2**. Pendiente e intersección para el ajuste de regresión a los datos pulmonares.

# **Valores Ajustados y Residuos**

Conceptos importantes en el análisis de regresión son los `valores ajustados (las predicciones)` y `los residuos (errores de predicción)`. En general, los datos no se ajustan exactamente a una línea, por lo que la ecuación de regresión debería incluir un término de error explícito $e_i$:

$$Y_i = b_0 + b_1 X_i + e_i$$

Los valores ajustados, también conocidos como valores predichos, se suelen denotar como $\hat{Y_i}$. Estos se obtienen mediante la fórmula:

$$\hat{Y_i} = \hat{b_0} + \hat{b_1} X_i$$

La notación $\hat{b_0}$ y $\hat{b_1}$ indica que los coeficientes son estimados en lugar de conocidos.

Cuando ajustamos un modelo a datos, como un línea de regresión, queremos encontrar la mejor línea que se ajuste a esos datos. En otras palabras, queremos encontrar la línea que minimiza la distancia entre los puntos reales y la línea.
- Imagina que tienes una serie de puntos en un gráfico, y quieres dibujar una línea que se ajuste lo mejor posible a esos puntos, la `regresión de minimos cuadrados` es como encontrar la línea que hace que la suma de las distancias al cuadrado entre los puntos y la línea sea la más pequeña posible.
- Por ejemplo, si tenemos puntos que representan el tiempo de estudio de un estudiante y sus calificaciones en un examen, la regresión de mínimos cuadrados nos ayudaría a encontrar la línea que mejor se ajuste a estos puntos, lo que nos permitiría predecir las calificaciones basadas en el tiempo de estudio.
- En resumen, la regresión de mínimos cuadrados es simplemente encontrar la mejor línea que se ajuste a los datos al minimizar la distancia entre los puntos reales y la línea de regresión.

### Notación "Sombrero": Estimaciones versus Valores Conocidos
La notación del "sombrero" se utiliza para diferenciar entre estimaciones y valores conocidos. Entonces, el símbolo $\hat{b}$ ("b-sombrero") es una estimación del parámetro desconocido $b$. ¿Por qué los estadísticos diferencian entre la estimación y el valor real? La estimación tiene incertidumbre, mientras que el valor real es fijo.


Calculamos los `residuos` $\hat{e_i}$ restando los valores predichos de los datos originales:
$$\hat{e_i} = {Y_i} - \hat{Y_i}$$

En R, podemos obtener los valores ajustados y los residuos utilizando las funciones `predict` y `residuals`:

    fitted <- predict(model)
    resid <- residuals(model)

Con el modelo de `regresión lineal` de `scikit-learn`, utilizamos el método `predict` en los datos de entrenamiento para obtener los valores ajustados y, posteriormente, los residuos. Como veremos, este es un patrón general que siguen todos los modelos en `scikit-learn`:

```python
fitted = model.predict(lung[predictors])
residuals = lung[outcome] - fitted
```

# **Mínimos Cuadrados**

¿Cómo se ajusta el modelo a los datos? Cuando hay una relación clara, podrías imaginar ajustar la línea manualmente. En la práctica, la línea de regresión es la estimación que minimiza la suma de los valores residuales al cuadrado, también llamada `suma residual de cuadrados` o `RSS`:


Las estimaciones $b_0$ y $b_1$ son los valores que minimizan RSS.
$$RSS = \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2 $$
$$= \sum_{i=1}^{n} (Y_i - \hat{b_0} - \hat{b_1} X_i)^2$$

El método de minimizar la suma de los residuos al cuadrado se denomina regresión de mínimos cuadrados, o regresión de mínimos cuadrados ordinarios (OLS). A menudo se atribuye a Carl Friedrich Gauss, el matemático alemán, pero fue publicado por primera vez por el matemático francés Adrien-Marie Legendre en 1805. La regresión de mínimos cuadrados se puede calcular rápida y fácilmente con cualquier software estadístico estándar.

Históricamente, la conveniencia computacional es una razón para el amplio uso de los mínimos cuadrados en regresión. Con la llegada de big data, la velocidad computacional sigue siendo un factor importante. Los mínimos cuadrados, al igual que la media (ver "Mediana y Estimaciones Robustas" en la página 10), son sensibles a los valores atípicos, aunque esto tiende a ser un problema significativo solo en conjuntos de datos pequeños o de tamaño moderado. Consulta "Valores Atípicos" en la página 177 para obtener una discusión sobre valores atípicos en regresión.

- El método de `minimos cuadrados` se utiliza para ajustar una línea (o curva) a los datos de manera que minimica la diferencia entre los valores reales y los predichos por el modelo. Esto se logra calculando los valores que minimizan la suma de los cuadrados de los errores residuales, es decir, la diferencia entre los valores observados y los predichos por el modelo.
En la práctica, estos valores son $b_0$ y $b_1$, que son los `coeficientes de la ecuación de regresión lineal` (intercepto y pendiente, respectivamente) que minimizan la suma de los cuadrados de los residuos (RSS). Los `residuos` son simplemente las diferencias entre los `valores observados` $Y_i$ y los `valores predichos` $\hat{Y_i}$ por el modelo.

- La `regresión de minimos cuadrados (OLS)` es ampliamente utilizada debido a su conveniencia computacional y su capacidad para proporciona estimaciones rápidas y fáciles de interpretar. Aunque es sensible a los valore atípicos en los datos, este problema suele ser más significativo en conjuntos de datos pequeños o de tamaño moderado.

- En resumen, la regresión de mínimos cuadrados encuentra la mejor línea o curva que se ajusta a los datos al minimizar la diferencia entre los valores reales y los predichos, lo que la convierte en una herramienta útile para entender y predecir relaciones entre variables.

### Terminología de Regresión
    Cuando analistas e investigadores usan el término "regresión" por sí mismo, generalmente se están refiriendo a la regresión lineal; el enfoque suele estar en desarrollar un modelo lineal para explicar la relación entre las variables predictoras y una variable de resultado numérica. En su sentido estadístico formal, la regresión también incluye modelos no lineales que proporcionan una relación funcional entre las variables predictoras y las variables de resultado. En la comunidad de aprendizaje automático, el término a veces se usa de manera más laxa para referirse al uso de cualquier modelo predictivo que produce un resultado numérico predicho (en oposición a los métodos de clasificación que predicen un resultado binario o categórico).

# **Predicción versus Explicación** (Perfilación "Profiling")

Históricamente, un uso primario de la regresión era iluminar una supuesta relación lineal entre las variables predictoras y una variable de resultado. El objetivo era comprender una relación y explicarla utilizando los datos para los que se ajustó la regresión. En este caso, el enfoque principal está en la pendiente estimada de la ecuación de regresión, $\hat{b}$. Los economistas quieren conocer la relación entre el gasto del consumidor y el crecimiento del PIB. Los funcionarios de salud pública podrían querer entender si una campaña de información pública es efectiva para promover prácticas sexuales seguras. En estos casos, el enfoque no está en predecir casos individuales, sino en comprender la relación general entre las variables.

Con la llegada de big data, la `regresión` se utiliza ampliamente para formar un modelo para predecir resultados individuales para nuevos datos (es decir, un modelo predictivo) en lugar de explicar los datos existentes. En este caso, los principales elementos de interés son los valores ajustados $Y$. En marketing, la regresión se puede utilizar para predecir el cambio en los ingresos en respuesta al tamaño de una campaña publicitaria. Las universidades utilizan la regresión para predecir el GPA de los estudiantes en función de sus puntajes SAT.

Un `modelo de regresión` que se ajusta bien a los datos se configura de manera que los cambios en $X$ conducen a cambios en $Y$. Sin embargo, por sí sola, la ecuación de regresión no demuestra la dirección de la causalidad. Las conclusiones sobre la causalidad deben provenir de una comprensión más amplia sobre la relación. Por ejemplo, una ecuación de regresión podría mostrar una relación definida entre el número de clics en un anuncio web y el número de conversiones. Es nuestro conocimiento del proceso de marketing, no la ecuación de regresión, lo que nos lleva a la conclusión de que los clics en el anuncio conducen a las ventas, y no al revés.

### Ideas Clave
* La ecuación de regresión modela la relación entre una variable de respuesta \(Y\) y una variable predictora \(X\) como una línea.
* Un modelo de regresión produce valores ajustados y residuos: predicciones de la respuesta y los errores de las predicciones.
* Los modelos de regresión generalmente se ajustan mediante el método de mínimos cuadrados.
* La regresión se utiliza tanto para la predicción como para la explicación.

# Regresión Lineal Múltiple
Cuando hay múltiples predictores, la ecuación simplemente se extiende para acomodarlos:

$$Y = b_0 + b_1X_1 + b_2X_2 + \ldots + b_pX_p + e$$

En lugar de una línea, ahora tenemos un modelo lineal. La relación entre cada coeficiente y su variable (característica) es lineal.

### Términos Clave para la Regresión Lineal Múltiple

**Error cuadrático medio raíz (RMSE)**
La raíz cuadrada del error cuadrático promedio de la regresión (esta es la métrica más utilizada para comparar modelos de regresión).
Sinónimo
RMSE

**Error estándar residual**
Lo mismo que el error cuadrático medio raíz, pero ajustado para los grados de libertad.
Sinónimo
RSE

**R-cuadrado**
La proporción de la varianza explicada por el modelo, de 0 a 1.
Sinónimos
coeficiente de determinación, R2

**Estadístico t**
El coeficiente para un predictor, dividido por el error estándar del coeficiente, proporcionando una métrica para comparar la importancia de las variables en el modelo. Ver "Pruebas t" en la página 110.

**Regresión ponderada**
Regresión con registros que tienen diferentes pesos.

Todos los demás conceptos en regresión lineal simple, como el ajuste por mínimos cuadrados y la definición de valores ajustados y residuos, se extienden al entorno de regresión lineal múltiple. Por ejemplo, los valores ajustados se expresan como:

$$\hat{Y_i} = \hat{b_0} + \hat{b_1}X_{1,i} + \hat{b_2}X_{2,i} + \ldots + \hat{b_p}X_{p,i}$$

# Evaluación del Modelo
La métrica de rendimiento más importante desde una perspectiva de ciencia de datos es el error cuadrático medio raíz, o RMSE. RMSE es la raíz cuadrada del error promedio al cuadrado en los valores \(y_i\) predichos:

$$ RMSE = \sqrt{\frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{n}} $$

Esto mide la precisión general del modelo y sirve como base para compararlo con otros modelos (incluyendo modelos ajustados mediante técnicas de aprendizaje automático). Similar a RMSE es el error estándar residual, o RSE. En este caso, tenemos $p$ predictores, y el RSE se expresa como:

$$RSE = \sqrt{\frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{n - p - 1}}$$

La única diferencia es que el denominador es los grados de libertad, en lugar del número de registros (ver "Grados de Libertad" en la página 116). En la práctica, para la regresión lineal, la diferencia entre RMSE y RSE es muy pequeña, especialmente para aplicaciones con grandes conjuntos de datos.

Otra métrica útil que verás en la salida del software es el coeficiente de determinación, también llamado estadístico R-cuadrado o R2. R-cuadrado varía de 0 a 1 y mide la proporción de variación en los datos que se explica en el modelo. Es útil principalmente en usos explicativos de la regresión donde deseas evaluar qué tan bien el modelo se ajusta a los datos. La fórmula para R2 es:

$$ R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$

El denominador es proporcional a la varianza de \(Y\). La salida de R también informa un R-cuadrado ajustado, que se ajusta a los grados de libertad, penalizando efectivamente la adición de más predictores a un modelo. Rara vez esto es significativamente diferente de R-cuadrado en regresión múltiple con conjuntos de datos grandes.

Junto con los coeficientes estimados, R y statsmodels informan el error estándar de los coeficientes (SE) y un estadístico t:

$$t_b = \frac{b}{SE_b}$$

El estadístico t, y su imagen especular, el valor p, miden en qué medida un coeficiente es "estadísticamente significativo", es decir, fuera del rango de lo que podría producir una disposición aleatoria de las variables predictoras y la variable objetivo. Cuanto mayor es el estadístico t (y menor es el valor p), más significativo es el predictor. Dado que la parsimonia es una característica valiosa del modelo, es útil tener una herramienta como esta para guiar la elección de variables a incluir como predictores (ver "Selección de Modelo y Regresión Escalonada" en la página 156).

**Nota:**

Además del estadístico t, R y otros paquetes a menudo informarán un valor p (Pr(>|t|) en la salida de R) y estadístico F. Los científicos de datos generalmente no se involucran demasiado en la interpretación de estas estadísticas ni en la cuestión de la significancia estadística. Los científicos de datos se centran principalmente en el estadístico t como una guía útil para decidir si incluir o no un predictor en un modelo. Altos estadísticos t (que van de la mano con valores p cercanos a 0) indican que un predictor debería mantenerse en un modelo, mientras que estadísticos t muy bajos indican que un predictor podría eliminarse. # Validación Cruzada
Las métricas clásicas de regresión estadística (R
2
, estadísticas F y valores p) son todas métricas "en muestra" (in-sample): se aplican a los mismos datos que se utilizaron para ajustar el modelo. Intuitivamente, se puede ver que tendría mucho sentido apartar algunos de los datos originales, no utilizarlos para ajustar el modelo y luego aplicar el modelo a los datos apartados (conjunto de prueba) para ver qué tan bien funciona. Normalmente, usarías la mayoría de los datos para ajustar el modelo y una porción más pequeña para probar el modelo.

Esta idea de validación "fuera de muestra" no es nueva, pero realmente no se consolidó hasta que los conjuntos de datos más grandes se volvieron más comunes. Con un conjunto de datos pequeño, los analistas típicamente quieren utilizar todos los datos y ajustar el mejor modelo posible.

Sin embargo, el uso de una muestra de retención te expone a cierta incertidumbre que surge simplemente por la variabilidad en la pequeña muestra de retención. ¿Qué tan diferente sería la evaluación si seleccionaras una muestra de retención diferente?

La validación cruzada extiende la idea de una muestra de retención a múltiples muestras de retención secuenciales. El algoritmo para la validación cruzada básica k-fold es el siguiente:

1. Reserva 1/k de los datos como una muestra de retención.
2. Entrena el modelo en los datos restantes.
3. Aplica (evalúa) el modelo en la muestra de retención 1/k y registra las métricas de evaluación del modelo necesarias.
4. Restaura el primer 1/k de los datos y reserva el siguiente 1/k (excluyendo cualquier registro que se haya seleccionado la primera vez).
5. Repite los pasos 2 y 3.
6. Repite hasta que cada registro se haya utilizado en la porción de retención.
7. Promedia o combina de alguna manera las métricas de evaluación del modelo.

La división de los datos en la muestra de entrenamiento y la muestra de retención también se llama "fold" (pliegue).

Agregar más variables no significa necesariamente que tengamos un modelo mejor. Los estadísticos utilizan el principio de la navaja de Occam para guiar la elección de un modelo: todas las cosas siendo iguales, se debe preferir un modelo más simple a uno más complicado.

Incluir variables adicionales siempre reduce el RMSE y aumenta el R2 para los datos de entrenamiento. Por lo tanto, estos no son apropiados para ayudar a guiar la elección del modelo. Un enfoque para incluir la complejidad del modelo es utilizar el R2 ajustado:

$$R_{adj}^2 = 1 - \frac{(1 - R^2) \cdot (n - 1)}{n - P - 1}$$

Aquí, $n$ es el número de registros y $P$ es el número de variables en el modelo.

En la década de 1970, Hirotugu Akaike, el eminente estadístico japonés, desarrolló una métrica llamada AIC (Criterios de Información de Akaike) que penaliza la adición de términos a un modelo. En el caso de la regresión, AIC tiene la forma:

$$AIC = 2P + n \log\left(\frac{RSS}{n}\right)$$

donde $P$ es el número de variables y $n$ es el número de registros. El objetivo es encontrar el modelo que minimiza el AIC; los modelos con $k$ variables adicionales son penalizados por $2k$.

### AIC, BIC y Mallows Cp

La fórmula para AIC puede parecer un poco misteriosa, pero de hecho está basada en resultados asintóticos en teoría de la información. Hay varias variantes de AIC:

- **AICc:** Una versión de AIC corregida para tamaños de muestra pequeños.
- **BIC o criterios de información bayesiana:** Similar a AIC, pero con una penalización más fuerte por incluir variables adicionales en el modelo.
- **Mallows Cp:** Una variante de AIC desarrollada por Colin Mallows.

Estas métricas se informan típicamente como métricas en muestra (es decir, en los datos de entrenamiento), y los científicos de datos que utilizan datos de retención para la evaluación del modelo no necesitan preocuparse por las diferencias entre ellas ni por la teoría subyacente detrás de ellas.

¿Cómo encontramos el modelo que minimiza AIC o maximiza \( R_{adj}^2 \)? Una forma es buscar entre todos los modelos posibles, un enfoque llamado regresión de todos los subconjuntos. Esto es computacionalmente costoso y no es factible para problemas con grandes conjuntos de datos y muchas variables. Una alternativa atractiva es utilizar la regresión escalonada (stepwise regression). Se puede comenzar con un modelo completo y sucesivamente eliminar variables que no contribuyan significativamente. Esto se llama eliminación hacia atrás (backward elimination). Alternativamente, se podría comenzar con un modelo constante y agregar sucesivamente variables (selección hacia adelante). Como tercera opción, también podemos agregar y eliminar predictores sucesivamente para encontrar un modelo que reduzca AIC o \( R_{adj}^2 \). Esto significa que la selección del modelo posiblemente esté sujeta al sobreajuste (ajuste al ruido en los datos) y puede no funcionar tan bien cuando se aplica a nuevos datos. Un enfoque común para evitar esto es utilizar la validación cruzada para validar los modelos. En la regresión lineal, el sobreajuste generalmente no es un problema importante, debido a la estructura global simple (lineal) impuesta en los datos. Sin embargo, para tipos de modelos más sofisticados, especialmente procedimientos iterativos que responden a la estructura local de los datos, la validación cruzada es una herramienta muy importante; consulta "Validación Cruzada" en la página 155 para obtener más detalles.

# Predicción Utilizando Regresión
El propósito principal de la regresión en ciencia de datos es la predicción. Esto es útil tenerlo en cuenta, ya que la regresión, siendo un método estadístico antiguo y establecido, viene con conceptos más relevantes para su papel tradicional como una herramienta para modelado explicativo que para la predicción.

### Términos Clave para la Predicción Utilizando Regresión

**Intervalo de predicción**
Un intervalo de incertidumbre alrededor de un valor predicho individual.

**Extrapolación**
Extensión de un modelo más allá del rango de los datos utilizados para ajustarlo.

### Los Peligros de la Extrapolación

Los modelos de regresión no deben utilizarse para la extrapolación más allá del rango de los datos (aunque se excluya el uso de la regresión para la predicción de series temporales). El modelo es válido solo para valores de predictor para los cuales los datos tienen suficientes valores (incluso en el caso de que haya datos suficientes, podría haber otros problemas, consulte "Diagnóstico de regresión" en la página 176). Como caso extremo, supongamos que el modelo_lm se utiliza para predecir el valor de un lote vacío de 5,000 pies cuadrados. En tal caso, todos los predictores relacionados con el edificio tendrían un valor de 0, y la ecuación de regresión arrojaría una predicción absurda de -521,900 + 5,000 × -0.0605 = -$522,202. ¿Por qué ocurrió esto? Los datos solo contienen parcelas con edificios; no hay registros correspondientes a terrenos vacíos. En consecuencia, el modelo no tiene información para indicarle cómo predecir el precio de venta para terrenos vacíos.

# Confiabilidad e Intervalos de Predicción

Gran parte de la estadística implica comprender y medir la variabilidad (incertidumbre). Las estadísticas t y los valores p informados en la salida de la regresión abordan esto de manera formal, lo que a veces es útil para la selección de variables (ver "Evaluación del Modelo" en la página 153). Métricas más útiles son los intervalos de confianza, que son intervalos de incertidumbre colocados alrededor de los coeficientes y predicciones de regresión. Una forma sencilla de entender esto es a través del bootstrap (consultar "El Bootstrap" en la página 61 para obtener más detalles sobre el procedimiento general del bootstrap).

Los intervalos de confianza más comunes en la salida de software son aquellos para los parámetros de regresión (coeficientes). Aquí hay un algoritmo de bootstrap para generar intervalos de confianza para los parámetros de regresión (coeficientes) para un conjunto de datos con P predictores y n registros (filas):

1. Considere cada fila (incluida la variable de resultado) como un "boleto" único y coloque todos los n boletos en una caja.
2. Extraiga un boleto al azar, registre los valores y devuélvalo a la caja.
3. Repita el paso 2 n veces; ahora tiene una muestra de bootstrap.
4. Ajuste una regresión a la muestra de bootstrap y registre los coeficientes estimados.
5. Repita los pasos 2 a 4, digamos, 1,000 veces.
6. Ahora tiene 1,000 valores de bootstrap para cada coeficiente; encuentre los percentiles apropiados para cada uno (por ejemplo, el 5º y el 95º para un intervalo de confianza del 90%).

Puede utilizar la función Boot en R para generar intervalos de confianza de bootstrap reales para los coeficientes, o simplemente puede utilizar los intervalos basados en fórmulas que son una salida rutinaria en R. El significado conceptual e interpretación son los mismos y no son de vital importancia para los científicos de datos, porque se refieren a los coeficientes de regresión. De mayor interés para los científicos de datos son los intervalos alrededor de los valores y predichos (Yi).

La incertidumbre alrededor de Yi proviene de dos fuentes:
- Incertidumbre sobre cuáles son las variables predictoras relevantes y sus coeficientes (ver el algoritmo de bootstrap anterior).
- Error adicional inherente en puntos de datos individuales.

El error individual de puntos de datos se puede pensar de la siguiente manera: incluso si supiéramos con certeza cuál era la ecuación de regresión (por ejemplo, si tuviéramos un número enorme de registros para ajustarla), los valores reales de resultado para un conjunto dado de valores predictores variarán. Por ejemplo, varias casas, cada una con 8 habitaciones, un lote de 6,500 pies cuadrados, 3 baños y un sótano, podrían tener valores diferentes. Podemos modelar este error individual con los residuos de los valores ajustados. El algoritmo de bootstrap para modelar tanto el error del modelo de regresión como el error de puntos de datos individuales sería el siguiente:

1. Tomar una muestra de bootstrap de los datos (detallado con mayor detalle anteriormente).
2. Ajustar la regresión y predecir el nuevo valor.
3. Tomar un solo residuo al azar del ajuste de regresión original, agregarlo al valor predicho y registrar el resultado.
4. Repetir los pasos 1 a 3, digamos, 1,000 veces.
5. Encontrar los percentiles 2.5 y 97.5 de los resultados.

### Ideas Clave
* La extrapolación más allá del rango de los datos puede llevar a errores.
* Los intervalos de confianza cuantifican la incertidumbre alrededor de los coeficientes de regresión.
* Los intervalos de predicción cuantifican la incertidumbre en predicciones individuales.
* La mayoría del software, incluido R, producirá intervalos de predicción e intervalos de confianza en la salida predeterminada o especificada, utilizando fórmulas.
* El bootstrap también se puede utilizar para producir intervalos de predicción e intervalos de confianza; la interpretación e idea son las mismas.

### Intervalo de Predicción o Intervalo de Confianza
Un intervalo de predicción se refiere a la incertidumbre alrededor de un valor único,
mientras que un intervalo de confianza se refiere a una media u otro estadístico calculado a partir de múltiples valores. Por lo tanto, un intervalo de predicción típicamente será mucho más amplio que un intervalo de confianza para el mismo valor.
Modelamos este error en el valor individual en el modelo bootstrap seleccionando un residuo individual para agregar al valor predicho.
¿Cuál deberías usar? Eso depende del contexto y del propósito del análisis, pero en general, los científicos de datos están interesados en predicciones individuales específicas, por lo que un intervalo de predicción sería más apropiado. Usar un intervalo de confianza cuando deberías estar usando un intervalo de predicción subestimará en gran medida la incertidumbre en un valor predicho dado.

# Variables de Factor en Regresión
Las variables de factor, también llamadas variables categóricas, toman un número limitado de valores discretos. Por ejemplo, el propósito de un préstamo puede ser "consolidación de deudas", "boda", "automóvil", y así sucesivamente. La variable binaria (sí/no), también llamada variable indicadora, es un caso especial de una variable de factor. La regresión requiere entradas numéricas, por lo que las variables de factor deben recodificarse para usarlas en el modelo. El enfoque más común es convertir una variable en un conjunto de variables binarias dummy.

### Términos clave relacionados con variables de factor

1. **Variables Dummy:**
   Variables binarias 0-1 derivadas al recodificar datos de factor para su uso en regresión y otros modelos.

2. **Codificación de Referencia:**
   El tipo más común de codificación utilizado por estadísticos, en el cual un nivel de un factor se utiliza como referencia y otros factores se comparan con ese nivel.

   Sinónimo: Codificación de tratamiento

3. **Codificador Uno Caliente:**
   Un tipo común de codificación utilizado en la comunidad de aprendizaje automático, en el cual se retienen todos los niveles del factor. Aunque es útil para ciertos algoritmos de aprendizaje automático, este enfoque no es apropiado para regresión lineal múltiple.

4. **Codificación de Desviación:**
   Un tipo de codificación que compara cada nivel con la media general en lugar del nivel de referencia.

   Sinónimo: Contrastes de suma

# Interpretar la Ecuación de Regresión

En ciencia de datos, el uso más importante de la regresión es predecir alguna variable dependiente (resultado). Sin embargo, en algunos casos, obtener información de la ecuación en sí misma para entender la naturaleza de la relación entre los predictores y el resultado puede ser valioso. Esta sección proporciona orientación sobre cómo examinar la ecuación de regresión e interpretarla.

### Términos Clave para Interpretar la Ecuación de Regresión

1. **Variables Correlacionadas:**
   Cuando las variables predictoras están altamente correlacionadas, es difícil interpretar los coeficientes individuales.

2. **Multicolinealidad:**
   Cuando las variables predictoras tienen correlación perfecta o casi perfecta, la regresión puede ser inestable o imposible de calcular.

   Sinónimo: colinealidad

3. **Variables de Confusión:**
   Un predictor importante que, cuando se omite, lleva a relaciones espurias en una ecuación de regresión.

4. **Efectos Principales:**
   La relación entre un predictor y la variable de resultado, independiente de otras variables.

5. **Interacciones:**
   Una relación interdependiente entre dos o más predictores y la respuesta.

# Multicolinealidad

Un caso extremo de variables correlacionadas produce multicolinealidad, una condición en la que hay redundancia entre las variables predictoras. La multicolinealidad perfecta ocurre cuando una variable predictora puede expresarse como una combinación lineal de otras. La multicolinealidad ocurre cuando:
- Una variable se incluye múltiples veces por error.
- Se crean P variables dummy en lugar de P – 1 a partir de una variable de factor (ver "Variables de Factor en Regresión" en la página 163).
- Dos variables están casi perfectamente correlacionadas entre sí.

La multicolinealidad en la regresión debe abordarse: las variables deben eliminarse hasta que la multicolinealidad desaparezca. Una regresión no tiene una solución bien definida en presencia de multicolinealidad perfecta. Muchos paquetes de software, incluyendo R y Python, manejan automáticamente ciertos tipos de multicolinealidad. Por ejemplo, si SqFtTotLiving se incluye dos veces en la regresión de los datos de vivienda, los resultados son los mismos que para el modelo house_lm. En el caso de multicolinealidad no perfecta, el software puede obtener una solución, pero los resultados pueden ser inestables.

**Nota:**

La multicolinealidad no es tan problemática para métodos de regresión no lineales como árboles, clustering y vecinos más cercanos, y en tales métodos puede ser aconsejable retener P variables dummy (en lugar de P – 1). Dicho esto, incluso en esos métodos, la no redundancia en las variables predictoras sigue siendo una virtud.

# Diagnóstico de Regresión

En modelado explicativo (es decir, en un contexto de investigación), se llevan a cabo diversos pasos, además de las métricas mencionadas anteriormente (consulte "Evaluación del Modelo" en la página 153), para evaluar qué tan bien se ajusta el modelo a los datos; la mayoría se basa en el análisis de los residuos. Estos pasos no abordan directamente la precisión predictiva, pero pueden proporcionar información útil en un entorno predictivo.

### Términos Clave para el Diagnóstico de Regresión:

- **Residuos Estandarizados**
  Residuos divididos por el error estándar de los residuos.

- **Valores Atípicos**
  Registros (o valores de resultado) que están distantes del resto de los datos (o del resultado predicho).

- **Valor Influyente**
  Un valor o registro cuya presencia o ausencia tiene un gran impacto en la ecuación de regresión.

- **Apalancamiento**
  El grado de influencia que tiene un solo registro en una ecuación de regresión.
  Sinónimo: valor-hat (hat-value).

- **Residuos no Normales**
  Residuos no distribuidos normalmente que pueden invalidar algunos requisitos técnicos de la regresión, pero generalmente no son una preocupación en la ciencia de datos.

- **Heterocedasticidad**
  Cuando algunos rangos del resultado experimentan residuos con mayor varianza (puede indicar la falta de algún predictor en la ecuación).

- **Gráficos de Residuos Parciales**
  Un gráfico diagnóstico para iluminar la relación entre la variable de resultado y un solo predictor.
  Sinónimo: gráfico de variables adicionadas.

# Outliers

En términos generales, un valor extremo, también llamado valor atípico u outlier, es aquel que está alejado de la mayoría de las otras observaciones. Así como los valores atípicos deben manejarse para las estimaciones de ubicación y variabilidad (consulte "Estimaciones de Ubicación" en la página 7 y "Estimaciones de Variabilidad" en la página 13), los outliers pueden causar problemas en modelos de regresión. En regresión, un outlier es un registro cuyo valor real de y está alejado del valor predicho. Puedes detectar outliers examinando el residuo estandarizado, que es el residuo dividido por el error estándar de los residuos.

No existe una teoría estadística que separe outliers de no outliers. En cambio, existen reglas generales (arbitrarias) sobre cuán alejada de la mayoría de los datos debe estar una observación para ser llamada un outlier. Por ejemplo, con el diagrama de caja (boxplot), los outliers son aquellos puntos de datos que están demasiado por encima o por debajo de los límites de la caja (consulte "Percentiles y Boxplots" en la página 20), donde "demasiado" significa "más de 1.5 veces el rango intercuartílico". En regresión, el residuo estandarizado es la métrica que se utiliza típicamente para determinar si un registro se clasifica como un outlier. Los residuos estandarizados se pueden interpretar como "la cantidad de desviaciones estándar lejos de la línea de regresión".

# Valores Influyentes

Un valor cuya ausencia cambiaría significativamente la ecuación de regresión se denomina observación influyente. En regresión, tal valor no necesariamente está asociado con un residuo grande. Como ejemplo, considera las líneas de regresión en la Figura 4-5. La línea sólida corresponde a la regresión con todos los datos, mientras que la línea discontinua corresponde a la regresión con el punto en la esquina superior derecha eliminado. Claramente, ese valor de datos tiene una gran influencia en la regresión aunque no esté asociado con un valor atípico grande (de la regresión completa). Este valor de datos se considera tener una alta influencia en la regresión.

Además de los residuos estandarizados (ver "Valores Atípicos" en la página 177), los estadísticos han desarrollado varias métricas para determinar la influencia de un solo registro en una regresión. Una medida común de la influencia es el valor-hat (hat-value); valores por encima de $2 \cdot \frac{P + 1}{n}$ indican un valor de datos con alta influencia.

# Heterocedasticidad, No Normalidad y Errores Correlacionados

Los estadísticos prestan considerable atención a la distribución de los residuos. Resulta que los mínimos cuadrados ordinarios (consulte "Mínimos Cuadrados" en la página 148) son insesgados y, en algunos casos, son el estimador "óptimo" bajo una amplia gama de supuestos de distribución. Esto significa que, en la mayoría de los problemas, los científicos de datos no necesitan preocuparse demasiado por la distribución de los residuos.

La distribución de los residuos es relevante principalmente para la validez de la inferencia estadística formal (pruebas de hipótesis y valores p), que tiene una importancia mínima para los científicos de datos preocupados principalmente por la precisión predictiva. Errores distribuidos normalmente son una señal de que el modelo es completo; errores que no están distribuidos normalmente indican que el modelo puede estar omitiendo algo. Para que la inferencia formal sea completamente válida, se asume que los residuos están distribuidos normalmente, tienen la misma varianza y son independientes.

Un área donde esto puede preocupar a los científicos de datos es el cálculo estándar de intervalos de confianza para valores predichos, que se basan en supuestos sobre los residuos (ver "Intervalos de Confianza y Predicción" en la página 161).

La heterocedasticidad es la falta de varianza residual constante a lo largo del rango de los valores predichos. En otras palabras, los errores son mayores para algunas partes del rango que para otras. Visualizar los datos es una forma conveniente de analizar los residuos.

Los estadísticos también pueden verificar la suposición de que los errores son independientes. Esto es especialmente cierto para datos recopilados a lo largo del tiempo o el espacio. La estadística de Durbin-Watson se puede utilizar para detectar si hay autocorrelación significativa en una regresión que involucra datos de series temporales. Consulta "Practical Time Series Forecasting with R, 2nd ed." de Galit Shmueli y Kenneth Lichtendahl (Axelrod Schnall, 2018) para obtener más información sobre cómo incorporar información de autocorrelación en modelos de regresión para datos de series temporales. Si el objetivo son pronósticos a más largo plazo o modelos explicativos, los datos autocorrelacionados en exceso a nivel micro pueden distraer. En ese caso, el suavizado o la recopilación de datos menos detallada desde el principio pueden ser apropiados.

Aunque una regresión pueda violar una de las suposiciones de distribución, ¿deberíamos preocuparnos? En la mayoría de los casos en la ciencia de datos, el interés se centra principalmente en la precisión predictiva, por lo que podría ser útil revisar la heterocedasticidad. Podrías descubrir que hay alguna señal en los datos que tu modelo no ha capturado. Sin embargo, satisfacer las suposiciones de distribución simplemente con el fin de validar la inferencia estadística formal (valores p, estadísticas F, etc.) no es tan importante para el científico de datos.

# Gráficos de Residuos Parciales y No Linealidad

Los gráficos de residuos parciales son una forma de visualizar cuán bien la ajuste estimado explica la relación entre un predictor y el resultado. La idea básica de un gráfico de residuos parciales es aislar la relación entre una variable predictora y la respuesta, teniendo en cuenta todas las demás variables predictoras. Un residuo parcial podría considerarse como un valor de "resultado sintético", que combina la predicción basada en una única variable predictora con el residuo real de la ecuación de regresión completa. Un residuo parcial para la variable predictora $Xi$ es el residuo ordinario más el término de regresión asociado con $Xi$:

$$\text{Residuo Parcial} = \text{Residuo} + \hat{b_i} \cdot X_i$$

# Regresión Polinómica y de Spline

La relación entre la respuesta y una variable predictora no necesariamente es lineal. La respuesta a la dosis de un medicamento a menudo no es lineal: duplicar la dosis generalmente no conduce a una respuesta duplicada. La demanda de un producto no es una función lineal de los dólares gastados en marketing; en algún momento, es probable que la demanda esté saturada. Hay muchas formas en que la regresión se puede ampliar para capturar estos efectos no lineales.

### Términos Clave para Regresión No Lineal

**Regresión polinómica**
Añade términos polinómicos (cuadrados, cúbicos, etc.) a una regresión.

**Regresión de spline**
Ajuste de una curva suave con una serie de segmentos polinómicos.

**Nodos**
Valores que separan los segmentos de spline.

**Modelos aditivos generalizados**
Modelos de spline con selección automática de nodos.
Sinónimo
GAM (Modelos Aditivos Generalizados, por sus siglas en inglés)

### Regresión No Lineal

Cuando los estadísticos hablan de regresión no lineal, se refieren a modelos que no pueden ajustarse mediante mínimos cuadrados. ¿Qué tipo de modelos son no lineales? Básicamente, todos los modelos en los que la respuesta no se puede expresar como una combinación lineal de los predictores o alguna transformación de los predictores. Los modelos de regresión no lineal son más difíciles y computacionalmente más intensivos de ajustar, ya que requieren optimización numérica. Por esta razón, generalmente se prefiere utilizar un modelo lineal si es posible.

# Splines
La regresión polinómica captura solo cierta cantidad de curvatura en una relación no lineal. Agregar términos de orden superior, como un polinomio cuártico cúbico, a menudo conduce a una "irregularidad" no deseada en la ecuación de regresión. Un enfoque alternativo, y a menudo superior, para modelar relaciones no lineales es utilizar splines. Los splines proporcionan una manera de interpolar suavemente entre puntos fijos. Los splines fueron utilizados originalmente por dibujantes para trazar una curva suave, especialmente en la construcción de barcos y aviones. Los splines se creaban doblando una pieza delgada de madera utilizando pesos, conocidos como "patos"; ver la Figura 4-11.

**Figura 4-11**. Los splines fueron creados originalmente utilizando madera flexible y "patos" y se usaban como herramienta de dibujo para ajustar curvas (foto cortesía de Bob Perry).

La definición técnica de un spline es una serie de polinomios continuos por partes. Fueron desarrollados por primera vez durante la Segunda Guerra Mundial en los Campos de Pruebas de Aberdeen en Estados Unidos por I. J. Schoenberg, un matemático rumano. Las piezas polinómicas están conectadas suavemente en una serie de puntos fijos en una variable predictora, denominados nudos. La formulación de los splines es mucho más complicada que la regresión polinómica; generalmente, el software estadístico maneja los detalles del ajuste de un spline. El paquete R splines incluye la función bs para crear un término de b-spline (spline de base) en un modelo de regresión. ### Ideas Clave:

# Resumen:
Quizás ningún otro método estadístico ha visto un uso mayor a lo largo de los años que la regresión, el proceso de establecer una relación entre múltiples variables predictoras y una variable de resultado. La forma fundamental es lineal: cada variable predictora tiene un coeficiente que describe una relación lineal entre el predictor y el resultado. Formas más avanzadas de regresión, como la regresión polinómica y de splines, permiten que la relación sea no lineal. En la estadística clásica, el énfasis está en encontrar un buen ajuste a los datos observados para explicar o describir algún fenómeno, y la fuerza de este ajuste es evaluada mediante métricas tradicionales de muestra para evaluar el modelo.

En ciencia de datos, por otro lado, el objetivo suele ser predecir valores para nuevos datos, por lo que se utilizan métricas basadas en la precisión predictiva para datos fuera de muestra. Se utilizan métodos de selección de variables para reducir la dimensionalidad y crear modelos más compactos.
